```
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_iris

# Load a sample dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Initialize and train two different models
model1 = LogisticRegression(max_iter=1000)
model2 = DecisionTreeClassifier()

model1.fit(X_train, y_train)
model2.fit(X_train, y_test)

# Make predictions on the test set
y_pred1 = model1.predict(X_test)
y_pred2 = model2.predict(X_test)

# Evaluate the models
accuracy1 = accuracy_score(y_test, y_pred1)
accuracy2 = accuracy_score(y_test, y_pred2)

print(f"Accuracy of Logistic Regression: {accuracy1}")
print(f"Accuracy of Decision Tree: {accuracy2}")

# You can also print classification reports for more detailed metrics
print("\nClassification Report for Logistic Regression:")
print(classification_report(y_test, y_pred1))

print("\nClassification Report for Decision Tree:")
print(classification_report(y_test, y_pred2))
```
